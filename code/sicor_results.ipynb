```python
# Function to calculate entropy
def entropy(series):
    value_counts = series.value_counts(normalize=True)
    return -np.sum(value_counts * np.log2(value_counts + 1e-10))  # Add small constant to avoid log(0)

# Discretize ContractTotalValue into quartiles
df['ContractTotalValue_Binned'] = pd.qcut(df['ContractTotalValue'], q=4, duplicates='drop')

# Calculate base entropy of the target (binned ContractTotalValue)
base_entropy = entropy(df['ContractTotalValue_Binned'])
print(f"Base Entropy of ContractTotalValue: {base_entropy:.4f}")

# Attributes to evaluate (excluding the target and binned column)
attributes = ['StateCode', 'MunicipalityCode', 'MonthIssued', 'ProgramCode', 'SubProgramCode',
              'ResourceSourceCode', 'EconomicActivity', 'FundingQuantity', 'FundingValue',
              'InvestmentQuantity', 'InvestmentValue', 'CommercializationQuantity',
              'CommercializationValue', 'IndustrializationQuantity', 'IndustrializationValue',
              'AllocatedAreaInvestiment', 'AllocatedAreaFunding', 'FinancialInstitutionCNPJ']

# Calculate information gain for each attribute
info_gains = {}
for attr in attributes:
    # Group by attribute and calculate conditional entropy
    grouped = df.groupby(attr)['ContractTotalValue_Binned'].apply(entropy)
    conditional_entropy = (grouped * df.groupby(attr).size() / len(df)).sum()
    info_gain = base_entropy - conditional_entropy
    info_gains[attr] = info_gain
    print(f"Information Gain for {attr}: {info_gain:.4f}")

# Identify attribute with maximum information gain
max_attr = max(info_gains, key=info_gains.get)
max_gain = info_gains[max_attr]

print(f"\nAttribute with Greatest Information Gain: {max_attr}")
print(f"Information Gain Value: {max_gain:.4f}")
```

```python
import pandas as pd

# Load the dataset
file_path = 'sicor_municipal_contracts.csv'
data = pd.read_csv(file_path)

# Ensure relevant columns are numeric
data['industrialization_value'] = pd.to_numeric(data['industrialization_value'], errors='coerce')
data['commercialization_value'] = pd.to_numeric(data['commercialization_value'], errors='coerce')

# Drop rows with missing values in relevant columns
data = data.dropna(subset=['industrialization_value', 'commercialization_value', 'contract_state'])

# Calculate total contract value
data['total_value'] = data['industrialization_value'] + data['commercialization_value']

# Group by contract state
grouped = data.groupby('contract_state')

# Calculate weighted probability
results = []
for state, group in grouped:
    # Total weighted sum for the state
    total_weight = group['total_value'].sum()

    # Weighted sum where industrialization > commercialization
    weighted_sum = group.loc[group['industrialization_value'] > group['commercialization_value'], 'total_value'].sum()

    # Probability
    probability = weighted_sum / total_weight if total_weight > 0 else 0

    results.append({'contract_state': state, 'probability': probability})

# Convert results to a DataFrame
results_df = pd.DataFrame(results)

# Display the results
print("Probability by Contract State:")
print(results_df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


gpt = pd.DataFrame({
    'Correct': [9, 8, 5 , 7],
    'Incorrect': [0, 1, 4 , 2],
    'Warning': [0, 1 , 1 , 0],
    'Interruption': [0, 0 , 2, 1]
}, index=['Descriptive' , 'Diagnostic' , 'Predictive' , "Prescriptive"])

grok = pd.DataFrame({
    'Correct': [9, 8, 5 , 7],
    'Incorrect': [0, 1, 4 , 2],
    'Warning': [0, 0 , 1 , 1],
    'Interruption': [0, 0 , 0, 0]
}, index=['Descriptive' , 'Diagnostic', 'Predictive' , "Prescriptive"])

qwen = pd.DataFrame({
    'Correct': [2, 0, 0, 0],
    'Incorrect': [7, 9, 9, 9],
    'Warning': [0 , 0 , 0 , 0],
    'Interruption': [7, 8, 9, 9]
}, index=['Descriptive' , 'Diagnostic' , 'Predictive' , "Prescriptive"])

# Lista com os DataFrames e nomes das bases
bases = [(gpt, "GPT analytcs"), (grok, "GROK 3"), (qwen, "QWEN 2.5 max")]

# Cores para as categorias
colors = {"Correct": "#77DD77", "Incorrect": "#FF6961", "Warning": "#f8fc5b", "Interruption": "#ffc05b"}

# Criando os 3 grficos
fig, axes = plt.subplots(1, 3, figsize=(18, 6), sharey=True)

for ax, (df, title) in zip(axes, bases):
    categories = df.index
    x = np.arange(len(categories))  # Posies das barras no eixo X
    width = 0.4  # Largura das barras

    # Criando barras empilhadas separadas em 2 grupos
    ax.bar(x - width/2, df['Correct'], color=colors["Correct"], width=width)
    ax.bar(x - width/2, df['Incorrect'], bottom=df['Correct'], color=colors["Incorrect"], width=width)

    ax.bar(x + width/2, df['Warning'], color=colors["Warning"], width=width)
    ax.bar(x + width/2, df['Interruption'], bottom=df['Warning'], color=colors["Interruption"], width=width)

    # Adicionando valores dentro das barras
    for i, category in enumerate(categories):
        if df['Correct'][category] > 0:
            ax.text(i - width/2, df['Correct'][category] / 2, str(df['Correct'][category]), ha='center', va='center', color='black')
        if df['Incorrect'][category] > 0:
            ax.text(i - width/2, df['Correct'][category] + df['Incorrect'][category] / 2, str(df['Incorrect'][category]), ha='center', va='center', color='black')
        if df['Warning'][category] > 0:
            ax.text(i + width/2, df['Warning'][category] / 2, str(df['Warning'][category]), ha='center', va='center', color='black')
        if df['Interruption'][category] > 0:
            ax.text(i + width/2, df['Warning'][category] + df['Interruption'][category] / 2, str(df['Interruption'][category]), ha='center', va='center', color='black')

    ax.set_xticks(x)
    ax.set_xticklabels(categories, rotation=45)
    ax.set_title(title)
    ax.set_ylim(0, 10)

# Criando legenda ao lado direito do ltimo plot
patches = [plt.Rectangle((0, 0), 1, 1, color=colors[label]) for label in colors]
axes[-1].legend(patches, colors.keys(),
               loc='center left',
               bbox_to_anchor=(1.05, 0.5),
               frameon=True,
               edgecolor='black',
               facecolor='white',
               title="Categories")

# Ajustando o layout para acomodar a legenda
plt.tight_layout(rect=[0, 0, 0.95, 1])  # Reduz o espao direito para 95% para caber a legenda
plt.show()

```
